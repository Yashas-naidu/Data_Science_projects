```python
import pandas as pd #used in working on csv file
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
tracks=pd.read_csv("C:\\Users\\swapn\\Data Science\\tracks.csv")
```

```python
#printing least popular artists in ascending order
least=tracks.sort_values("popularity",ascending=True).head(10)#head prints 10 values
least
```

```python
tracks.describe().transpose() #gives all statistics data
```

```python
#top 10 popular songs greater than 90
popular=tracks.query("popularity>90",inplace=False).sort_values('popularity',ascending=False)
popular[:10] 
#inplace is used so that original data set is not affected
```

```python
tracks.set_index("release_date",inplace=True)
tracks.index=pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks[["artists"]].iloc[4] #artist present in 4th index "indexing starts from 0"
```

```python
tracks.popularity.head(10)
```

```python
df=tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=tracks.sample(int(0.004*len(tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="popularity vs acousticness correlation")
```
